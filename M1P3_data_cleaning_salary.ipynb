* drop columns that are mostly 0/ empty

```python
# M1: salaries.drop(salaries.columns[[2, 3, 3]], axis = 0)
# M2: salaries=salaries.drop(columns=(['rowNumber', 'Masters_Degree','Bachelors_Degree','Doctorate_Degree','Highschool','Some_College', 'Race_Asian' ,'Race_White','Race_Two_Or_More','Race_Black','Race_Hispanic','Race']), axis = 0)
# M3 - using iloc
salaries = salaries.drop(salaries.iloc[:, 16:27], axis = 1)
```

* drop rows that are null with insignicantly small amount of Nulls (less than 200 rows out of 62K rows, excluding tag which is same as Title

```python
salaries = salaries.dropna(subset=['company','level','title','totalyearlycompensation','location','basesalary','stockgrantvalue','bonus'], axis=0)
```

* checking for duplicated strings/ name & standardize them

```python
salaries['company'] = salaries['company'].astype('string')
```

```python
salaries['company'] = salaries['company'].apply(str.upper)
salaries['company'].str.strip()
salaries['level'] = salaries['level'].apply(str.upper)
salaries['title'] = salaries['title'].apply(str.upper)

salaries = salaries.sort_values(by=['company'], ascending = True)
list(salaries.company.unique())
```

```python
# salaries.company.value_counts().to_csv("./company_counts.csv")
```

```python
# pd.set_option('display.max_rows', None)
salaries.company.value_counts()
```

* replacing mis-splelt company names

```python
salaries['company'] = salaries['company'].replace(['AMAZON WEB SERVICES','AMAZON.COM','AMZON'],'AMAZON')
salaries['company'] = salaries['company'].replace(['APPLE INC.'],'APPLE')
salaries['company'] = salaries['company'].replace(['MICROSOFT CORPORATION'],'MICROSOFT')
salaries['company'] = salaries['company'].replace(['Â GOOGLE','GOOGLE LLC','GOOGLE '],'GOOGLE')
```
